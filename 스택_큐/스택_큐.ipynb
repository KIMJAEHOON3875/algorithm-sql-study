# 스택
### 가장 마지막에 넣은 데이터가 가장 먼저 나오는 방식
### Push: 데이터를 맨 위에 쌓기 (aapend)
### Pop: 맨 위의 데이터를 꺼내기 (pop)

```python
# 1. 스택 초기화
stack = []

# 2. 데이터 넣기 (Push)
stack.append(10)
stack.append(20)
stack.append(30)
print(f"현재 스택: {stack}")

# 3. 데이터 꺼내기 (Pop)
top = stack.pop()
print(f"꺼낸 값:{top}")
print(f"남은 스택:{stack}")

# 4. 맨 위의 값만 확인하기 (Peek)
if stack:
    print(f"맨 위의 값: {stack[-1]}")
```

# 큐
### 먼저 온 사람이 먼저 나간다
### Enqueue: 줄 맨 끝에 서기 (append)
### Dequeue: 줄 맨 앞사람이 나가기 (popleft)
- 앞에서 빼기: popleft()
- 앞에 넣기: appendleft()
- 뒤에서 빼기: pop()
- 뒤에 넣기: append()

```python
from collections import deque

# 1. 큐 초기화
queue = deque()

# 2. 데이터 넣기 (Enqueue)
queue.append("A")
queue.append("B")
queue.append("C")
print(f"현재 큐 : {list(queue)}") 

# 3. 데이터 꺼내기 (Dequeque)
first = queue.popleft() # 맨 왼쪽에서 꺼내기
print(f"나간 사람: {first}")
print(f"남은 큐: {list(queue)}")

# 4. 맨 앞사람 확인하기
if queue:
    print(f"현재 1번 대기자: {queue[0]}")
```
